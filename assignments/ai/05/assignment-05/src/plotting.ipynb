```python
%matplotlib inline
import timeit
import os
import csv
import matplotlib.pyplot as plt 
from helper import City, hill_climb
```

```python
def read_data(file_name):
    working_directory = os.path.dirname(os.path.abspath(''))

    file_path = os.path.join(working_directory, "src/data/", file_name)

    with open(file_path) as file:
        reader = csv.DictReader(file)
        cities = [City(row["name"], row["longitude"], row["latitude"]) for row in reader]
        
    return cities

partial_cities = read_data("49_cities.txt")
full_cities = read_data("cities_full.txt")
```

```python
shortest_distance, shortest_sequence, distances = hill_climb(partial_cities, generate_report=True)
plt.plot(list(range(len(distances))), distances)
plt.title("Simple Hillclimbing for 49 cities")

plt.axis([-5000,105000, 0,5000]) # set axis 
plt.xlabel("Iteration")
plt.ylabel("Total euclidean distance")

plt.show()
```

```python
shortest_distance, shortest_sequence, distances = hill_climb(full_cities, generate_report=True)
plt.plot(list(range(len(distances))), distances)
plt.title("Simple Hillclimbing for all cities")

plt.axis([-5000,105000, 0,80000]) # set axis 
plt.xlabel("Iteration")
plt.ylabel("Total euclidean distance")

plt.show()
```

```python
shortest_distance, shortest_sequence, distances = hill_climb(partial_cities, generate_report=True, steepest_hill=True)
plt.plot(list(range(len(distances))), distances)
plt.title("Steep Hillclimbing for 49 cities")

plt.axis([-5000,105000, 0,5000]) # set axis 
plt.xlabel("Iteration")
plt.ylabel("Total euclidean distance")

plt.show()
```

```python
shortest_distance, shortest_sequence, distances = hill_climb(full_cities, generate_report=True, steepest_hill=True)
plt.plot(list(range(len(distances))), distances)
plt.title("Steep Hillclimbing for all cities")

plt.axis([-5000,105000, 0,80000]) # set axis 
plt.xlabel("Iteration")
plt.ylabel("Total euclidean distance")

plt.show()
```

```python
shortest_distance, shortest_sequence, distances = hill_climb(full_cities, generate_report=True, steepest_hill=True, pre_sort=True)
plt.plot(list(range(len(distances))), distances)
plt.title("Steep Hillclimbing for all cities")

plt.axis([-5000,105000, 0,5000]) # set axis 
plt.xlabel("Iteration")
plt.ylabel("Total euclidean distance")

plt.show()
```
